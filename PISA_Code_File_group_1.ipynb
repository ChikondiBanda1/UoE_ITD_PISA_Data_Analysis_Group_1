### Data preparation

Some of the variables are categorical in nature but have been saved as continuous. We need to change their `type` to `category`.

```python
# schooltype takes numerical values of 1,2,3
print(GB["SCHLTYPE"].value_counts())
```

```python
# fitting a normal regression using this independant/predictor variable is not correct because it is considered as continuous
from statsmodels.formula.api import ols
model_ss = ols('PV3MATH ~ SCHLTYPE', data=GB).fit()
model_ss.summary()
```

```python
# change the type to category with meaningful level names
GB['SCHLTYPE'] = GB['SCHLTYPE'].astype('category')
changes = {1:"Private Independent", 2:"Private Government-dependent", 3:"Public"}
GB["SCHLTYPE"] = GB["SCHLTYPE"].cat.rename_categories(changes)
print(GB["SCHLTYPE"].value_counts())
```

```python
# fit the regression model again- now there is a coefficient for each level of the categorical predictor
# (the first level is considered as the base line and doesn't have a coefficient).
model_ss = ols('PV3MATH ~ SCHLTYPE', data=GB).fit()
model_ss.summary()
```

```python
# do the samething for ST004D01T and save it as a new column named "gender"
GB['gender'] = GB['ST004D01T'].astype('category')
print(GB["gender"].value_counts())
```

```python
changes = {1:"female", 2:"male"}
GB['gender'] = GB['gender'].cat.rename_categories(changes)
print(GB['gender'].value_counts())
```

```python
# remove the `ST004D01T` column
GB.drop('ST004D01T', axis=1, inplace=True)
```
